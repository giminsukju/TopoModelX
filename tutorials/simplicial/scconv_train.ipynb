```python
import torch
import numpy as np
from sklearn.model_selection import train_test_split

from toponetx.datasets.mesh import shrec_16
from toponetx import SimplicialComplex
# from topomodelx.nn.simplicial.scconv_layer import SCConvLayer
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
shrec, _ = shrec_16(size="small")

shrec = {key: np.array(value) for key, value in shrec.items()}
x_0s = shrec["node_feat"]
x_1s = shrec["edge_feat"]
x_2s = shrec["face_feat"]

ys = shrec["label"]
simplexes = shrec["complexes"]
```

```python
i_complex = 0
print(
    f"The {i_complex}th simplicial complex has {x_0s[i_complex].shape[0]} nodes with features of dimension {x_0s[i_complex].shape[1]}."
)
print(
    f"The {i_complex}th simplicial complex has {x_1s[i_complex].shape[0]} edges with features of dimension {x_1s[i_complex].shape[1]}."
)
print(
    f"The {i_complex}th simplicial complex has {x_2s[i_complex].shape[0]} faces with features of dimension {x_2s[i_complex].shape[1]}."
)
```

```python
incidence_1_list = []
incidence_2_list = []
adjacency_upper_0_list = []
hodge_laplacian_1_list = []
adjacency_lower_2_list = []

for idx, simplex in enumerate(simplexes):
    print(simplex)
    print(simplex.is_triangular_mesh())
    
    simplex = simplex.to_cell_complex() # TODO: is this lifting necessary? 

    incidence_1 = simplex.incidence_matrix(rank=1, signed=True) # 0 -> 1, # TODO: signed should be true since SC?
    incidence_1 = torch.from_numpy(incidence_1.todense()).to_sparse() # TODO: why need this? 
    incidence_1_list.append(incidence_1)

    incidence_2 = simplex.incidence_matrix(rank=2, signed=True) # TODO: signed should be true since SC?
    incidence_2 = torch.from_numpy(incidence_2.todense()).to_sparse()
    incidence_2_list.append(incidence_2)

    adjacency_upper_0 = simplex.adjacency_matrix(rank=0) # TODO: check optional arguments
    adjacency_upper_0 = torch.from_numpy(adjacency_upper_0.todense()).to_sparse()
    adjacency_upper_0_list.append(adjacency_upper_0)

    hodge_laplacian_1 = simplex.hodge_laplacian_matrix(rank=1) # TODO: check optional arguments
    hodge_laplacian_1 = torch.from_numpy(hodge_laplacian_1.todense()).to_sparse() # TODO: still sparse?
    hodge_laplacian_1_list.append(hodge_laplacian_1_list)

    adjacency_lower_2 = simplex.coadjacency_matrix(rank=2) # TODO: check optional arguments
    adjacency_lower_2 = torch.from_numpy(adjacency_lower_2.todense()).to_sparse()
    adjacency_lower_2_list.append(adjacency_lower_2)
```
